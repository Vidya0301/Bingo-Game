```python
# Fixed list of industrial safety concepts
safety_concepts = [
    "PPE", "Fire Drill", "Lockout Tagout", "Safety Signage", "Hazard Symbols",
    "MSDS", "Machine Guarding", "Emergency Exit", "First Aid", "Safety Audit",
    "Eye Wash Station", "Spill Kit", "Fall Protection", "Ventilation", "Ergonomics",
    "Electrical Safety", "Confined Space", "Noise Control", "Fire Extinguisher", "Safe Lifting",
    "Risk Assessment", "Evacuation Plan", "Hard Hat", "Respirator", "Safety Culture"
]

# Create a fixed 5x5 Bingo card
def generate_fixed_bingo_card():
    card = [safety_concepts[i*5:(i+1)*5] for i in range(5)]
    return card

# Display the Bingo card
def print_card(card, marked):
    print("\nBINGO CARD:")
    for i in range(5):
        for j in range(5):
            value = card[i][j]
            if marked[i][j]:
                print(f"[X]".ljust(15), end=" ")
            else:
                print(value.ljust(15), end=" ")
        print()
    print()

# Check for Bingo
def check_bingo(marked):
    # Check rows
    for row in marked:
        if all(row): return True

    # Check columns
    for col in zip(*marked):
        if all(col): return True

    # Check diagonals
    if all(marked[i][i] for i in range(5)): return True
    if all(marked[i][4-i] for i in range(5)): return True

    return False

# Main game loop
def play_bingo():
    card = generate_fixed_bingo_card()
    marked = [[False]*5 for _ in range(5)]

    # Predefined calling order of concepts (can be sequential or specific)
    calling_order = safety_concepts  # same order as card to ensure a win

    print_card(card, marked)

    for item in calling_order:
        input(f"Press Enter to call the next item...")
        print(f"\nðŸ“¢ CALLED ITEM: {item}\n")

        # Mark the item if it's on the card
        for i in range(5):
            for j in range(5):
                if card[i][j] == item:
                    marked[i][j] = True

        print_card(card, marked)

        if check_bingo(marked):
            print("ðŸŽ‰ BINGO! You've completed a row, column, or diagonal with safety terms!")
            break

# Run the game
if __name__ == "__main__":
    print("ðŸ”’ Welcome to INDUSTRIAL SAFETY BINGO (Deterministic Version)!")
    play_bingo()
```

```python
import time
import pandas as pd
from IPython.display import clear_output

# Safety concepts (25 for 5x5 grid)
safety_concepts = [
    "PPE", "Fire Drill", "Lockout Tagout", "Safety Signage", "Hazard Symbols",
    "MSDS", "Machine Guarding", "Emergency Exit", "First Aid", "Safety Audit",
    "Eye Wash Station", "Spill Kit", "Fall Protection", "Ventilation", "Ergonomics",
    "Electrical Safety", "Confined Space", "Noise Control", "Fire Extinguisher", "Safe Lifting",
    "Risk Assessment", "Evacuation Plan", "Hard Hat", "Respirator", "Safety Culture"
]

# Split into 5x5 grid
def make_card():
    return [safety_concepts[i*5:(i+1)*5] for i in range(5)]

# Initialize cards and marks
p1_card = make_card()
p2_card = make_card()
p1_marks = [[False]*5 for _ in range(5)]
p2_marks = [[False]*5 for _ in range(5)]

# Check for Bingo
def is_bingo(marks):
    for i in range(5):
        if all(marks[i]): return True  # row
        if all(marks[j][i] for j in range(5)): return True  # col
    if all(marks[i][i] for i in range(5)): return True  # diag \
    if all(marks[i][4-i] for i in range(5)): return True  # diag /
    return False

# Mark the card
def mark_card(card, marks, item):
    for i in range(5):
        for j in range(5):
            if card[i][j] == item:
                marks[i][j] = True

# Display a card
def display_card(card, marks):
    display = []
    for i in range(5):
        row = []
        for j in range(5):
            if marks[i][j]:
                row.append(f"[{card[i][j]}]")
            else:
                row.append(card[i][j])
        display.append(row)
    return pd.DataFrame(display)

# Start Game
print("Starting Industrial Safety Bingo Game...\n")
time.sleep(2)

winner = None
for turn, item in enumerate(safety_concepts):
    clear_output(wait=True)
    print(f"Turn {turn+1}: Calling --> {item}\n")

    mark_card(p1_card, p1_marks, item)
    mark_card(p2_card, p2_marks, item)

    print("Player 1 Card:")
    display(display_card(p1_card, p1_marks))

    print("\nPlayer 2 Card:")
    display(display_card(p2_card, p2_marks))

    if is_bingo(p1_marks):
        winner = "Player 1"
        break
    elif is_bingo(p2_marks):
        winner = "Player 2"
        break

    time.sleep(2)

if winner:
    print(f"\nðŸŽ‰ {winner} has BINGO and wins the game!")
else:
    print("\nGame over. No winner.")

```

```python
import pandas as pd
from IPython.display import clear_output

# Safety concepts
safety_concepts = [
    "PPE", "Fire Drill", "Lockout Tagout", "Safety Signage", "Hazard Symbols",
    "MSDS", "Machine Guarding", "Emergency Exit", "First Aid", "Safety Audit",
    "Eye Wash Station", "Spill Kit", "Fall Protection", "Ventilation", "Ergonomics",
    "Electrical Safety", "Confined Space", "Noise Control", "Fire Extinguisher", "Safe Lifting",
    "Risk Assessment", "Evacuation Plan", "Hard Hat", "Respirator", "Safety Culture"
]

# Generate fixed 5x5 card from list
def make_card():
    return [safety_concepts[i*5:(i+1)*5] for i in range(5)]

# Check for Bingo
def is_bingo(marks):
    for i in range(5):
        if all(marks[i]): return True  # row
        if all(marks[j][i] for j in range(5)): return True  # column
    if all(marks[i][i] for i in range(5)): return True  # main diagonal
    if all(marks[i][4 - i] for i in range(5)): return True  # anti diagonal
    return False

# Mark card
def mark_card(card, marks, item):
    for i in range(5):
        for j in range(5):
            if card[i][j].lower() == item.lower():
                marks[i][j] = True

# Display card with marked items
def display_card(card, marks):
    display = []
    for i in range(5):
        row = []
        for j in range(5):
            word = card[i][j]
            row.append(f"[{word}]" if marks[i][j] else word)
        display.append(row)
    return pd.DataFrame(display)

# Initialize cards and states
p1_card = make_card()
p2_card = make_card()
p1_marks = [[False]*5 for _ in range(5)]
p2_marks = [[False]*5 for _ in range(5)]

called_items = []
winner = None

print("ðŸŽ® Industrial Safety Bingo Game - 2 Players")
print("âœ”ï¸ Enter one safety concept at a time (as written in the card).")
print("âŒ To quit early, type 'exit'.\n")

# Game loop
while not winner:
    called = input("ðŸ”” Enter called safety concept: ").strip()

    if called.lower() == "exit":
        print("\nâ—Game exited by user.")
        break

    if called not in safety_concepts:
        print("âš ï¸ Invalid concept. Try again.\n")
        continue

    if called in called_items:
        print("âš ï¸ Already called. Try another.\n")
        continue

    called_items.append(called)
    mark_card(p1_card, p1_marks, called)
    mark_card(p2_card, p2_marks, called)

    clear_output(wait=True)
    print(f"ðŸ”” Called: {called}\n")

    print("ðŸ§‘ Player 1 Card:")
    display(display_card(p1_card, p1_marks))

    print("\nðŸ§‘ Player 2 Card:")
    display(display_card(p2_card, p2_marks))

    if is_bingo(p1_marks):
        winner = "Player 1"
    elif is_bingo(p2_marks):
        winner = "Player 2"

if winner:
    print(f"\nðŸ† {winner} has BINGO! Congratulations!")
elif not winner and called.lower() != "exit":
    print("\nðŸš« No Bingo achieved.")
```
